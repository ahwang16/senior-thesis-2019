```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cluster, precision, recall = [], [], []
with open("kmeans_scores_2.txt", "r") as kmeans :
    next(kmeans)
    for line in kmeans :
        l = line.split('\t')
        cluster.append(int(l[0]))
        precision.append(float(l[1]))
        recall.append(float(l[2]))
```

```python
plt.scatter(cluster, precision)
plt.yticks(np.arange(0, 1.1, 0.1))
```

```python
plt.scatter(cluster, recall)
```
